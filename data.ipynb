```python
from PIL import Image
from matplotlib.pyplot import imshow
import pandas
import matplotlib.pylab as plt
import os
import glob
import skillsnetwork
```

```python
import requests, zipfile, io
import os

url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0321EN/data/images/concrete_crack_images_for_classification.zip"
output_dir = "data"

os.makedirs(output_dir, exist_ok=True)

# Download and extract
r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall(output_dir)

print("Dataset downloaded and extracted successfully.")
```

```python
negative_file_path=os.path.join("data","Negative")
negative_file_path
```

```python
os.listdir(negative_file_path)[0:3]
```

```python
negative_files=[os.path.join(negative_file_path,file) for file in  os.listdir(negative_file_path) if file.endswith(".jpg")]
negative_files.sort()
negative_files[0:3]
```

```python
positive_file_path=os.path.join("data","Positive")
positive_file_path
```

```python
positive_files=[os.path.join(positive_file_path,file) for file in os.listdir(positive_file_path) if file.endswith(".jpg")]
positive_files.sort()
positive_files[0:3]
```

```python
for i in range(3):
    ima=Image.open(positive_files[i])
    plt.imshow(ima)
    plt.title(f"{i}")
    plt.show()
```
